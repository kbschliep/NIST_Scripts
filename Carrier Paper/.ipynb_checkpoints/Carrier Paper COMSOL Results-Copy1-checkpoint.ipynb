```python
## Packages
'''Import all the necessary packages and modules'''

%matplotlib inline
import numpy as np # package for playing with matrices
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors
import sys, os # python modules for interacting with computer
from tkinter import filedialog # interface for opening a file explorer
from tkinter import * # * imports all files in tkinter 
import pandas as pd # useful for keeping cell structure and 2d data manipulation
import glob # helps find path 
```

```python
## Data Directory GUI
'''Getting data directory so data can easily be imported '''

root = Tk() # Tk() is a function in tkinter that opens a window
root.directory = filedialog.askdirectory() # opens explorer window so you can find the folder of choice
root.withdraw() # closes the tkinter window since it's unnecessary
oldcwd = os.getcwd() # saves old called working directory (place where data is drawn from) as oldcwd use os.chdir(oldcwd) to go back
os.chdir(root.directory) # sets new directory
newcwd = os.getcwd() # saves new directory name as newcwd
```

```python
## Load Data use for .csv data
''' Makes a list of all the files with the extension use * to '''

path = root.directory + '/*_50ohm_Reordered.csv' # Change '' as needed
a=0 # counter for looping through list starting at 0
file_list=glob.glob(path)
file_list
```

```python
## Import Data
''' Open data and set up for analysis in data_array matrix'''

num_files = np.size(file_list)-1 # number of files in the directory with the correct file type
file_shape = np.array(pd.read_csv(file_list[0], header=None).values.shape) # data file size
matrix_shape = np.append(file_shape, num_files) # size of matrix to hold all data
data_array = np.empty(matrix_shape) # builds empty matrix of size matrix_shape
a=0
for fname in file_list: #iterates through files along order of glob.glob(path)

    if fname==file_list[1]:
        continue
    else:
        data = pd.read_csv(fname, header=None) # uses pandas to open files and keep structure
        datapd = data.values # converts from dataframe to numpy array
#         print(a)
        data_array[:,:,a] = datapd # inserts each file as new layer of matrix along z axis (axis=2)
        a+=1 # increase counter to progress through list
```

```python
## Reorganize Data into Correct Order
'''From reorder into ascending er '''

X=data_array[:,0,0]; # All used the same x range
data_array_reorder=np.zeros(data_array.shape)
listr=[1,2,4,5,6,3]
a=0
for n in listr:
    data_array_reorder[:,:,int(n)-1]=data_array[:,:,a];# puts array in order of er ->2,3.8,4.4,6.7 10, 13.1
    
    a+=1

YS21=data_array_reorder[:,2::2,:].copy();# extracts S21 from each file - wrong for 3 aka er=6.7 data_array[:,:,2]
YS11=data_array_reorder[:,1::2,:].copy();# extracts S11 from each file - wrong for 3 aka er=6.7 data_array[:,:,2]
YS21[:,:,3]=data_array_reorder[:,13:,3]
YS11[:,:,3]=data_array_reorder[:,1:13,3]
```

```python
## Make image
'''Reorders data into a nice image'''
img_YS21=np.flipud(np.reshape(YS21,(YS21.shape[0],12*6),order='F')) # use [:178,:] if you want just 0-9 GHz
```

```python
## Plots data
''' Plots data as image'''

'''Setup plt properties'''
plt.rc('font', weight='bold')
plt.rcParams['axes.linewidth']=2
plt.rcParams['figure.figsize'] = (16,9)
plt.rcParams.update({'figure.autolayout': True})

'''Define colors'''
my_cmap = cm.jet_r
my_cmap.set_under('k', alpha=0)
my_cmap.set_over('k')
my_cmap1 = cm.binary_r
my_cmap1.set_under('k', alpha=0)

'''Create figure and frame'''
fig1 = plt.figure()
plt.tight_layout()

'''Add labels ontop of figure'''
fig1.text(0.94, 0.5,r'Insertion Loss, S$_{21}$ (dB)', ha='center', va='center', rotation='vertical', fontsize=36, fontweight='bold')
fig1.text(-0.02, 0.5,r'Input Signal Freq. (GHz)', ha='center', va='center', rotation='vertical', fontsize=36, fontweight='bold')
fig1.text(.45, -0.02,r'Conductor Width, S ($\mu$m)', ha='center', va='center', fontsize=36, fontweight='bold')

'''Adds additional frames to work with
Each ax is new frame that you can edit
'''
ax = fig1.add_subplot(111)
ax1=ax.twiny()
ax2=ax.twiny()

'''Setup all the ticks for different layers'''
ax.tick_params(axis='both', which='major', labelsize=24, width=2, length=6, labeltop=False)
ax.tick_params(axis='both', which='minor', labelsize=24, width=2, length=6, labeltop=False)

ax.xaxis.set_ticks(np.arange(.5/72,72/72,2/72))
label=np.tile([50,'','','','',550],6)
ax.xaxis.set_ticklabels(label,rotation=90)
           
ax1.set_xlabel(r'$\epsilon_r$', fontsize=44, fontweight='bold', labelpad=20)

ax1.tick_params(axis='both', which='major', labelsize=24, width=2, length=12, labeltop=False)
ax1.tick_params(axis='both', which='minor', labelsize=24, width=2, length=12, labeltop=False)

ax1.xaxis.set_ticks(np.arange(0,73,12)-.5)

loc1y=np.arange(0,319,317/8)
ax1.yaxis.set_ticks(loc1y)
label1y=np.arange(0,17,2)
ax1.yaxis.set_ticklabels(label1y[::-1])

ax2.tick_params(axis='both', which='major', labelsize=24, width=2, length=0)
ax2.tick_params(axis='both', which='minor', labelsize=24, width=2, length=0)

ax2.xaxis.set_ticks(np.arange(0,13,1))
label2=['',2,'',3.8,'',4.4,'',6.7,'',10,'',13.1]
ax2.xaxis.set_ticklabels(label2)

'''Shows the image'''
imgplot1=ax1.imshow(img_YS21, cmap=my_cmap, aspect='auto')

'''Generates colorbar'''
cbar=plt.colorbar(imgplot1)
cbar.ax.tick_params(labelsize=24, width=2, length=6)
```

```python
## Load Data use for .csv data
''' Open L calculation folder to get filenames'''
path1 = root1.directory + '/*.csv' # creates path to data set folder directory. Change .csv to file type as needed
a1=0 # counter for looping through list starting at 0
file_list1=glob.glob(path1)
file_list1
```

```python
## Load Data
''' Open L data and set up for analysis in matrix data_array1'''

num_files1 = np.size(file_list1) # number of files in the directory with the correct file type
file_shape1 = np.array(pd.read_csv(file_list1[0]).values.shape) # data file size
matrix_shape1 = np.append(file_shape1[0]+1, num_files1) # size of matrix to hold all data
data_array1 = np.empty(matrix_shape1) # builds empty matrix of size matrix_shape
a=0
lis=[3,1,0,5,4,2]
for fname in file_list1: #iterates through files along order of glob.glob(path)
    
        data = pd.read_csv(fname, header=None) # uses pandas to open files and keep structure
        datapd = data.values # converts from dataframe to numpy array
#         print(a)
        data_array1[:,lis[a]] = datapd.ravel() # inserts each file as new layer of matrix along z axis (axis=2)
        a+=1 # increase counter to progress through list
```

```python
## Average Data
'''Uses a function to compute average S21 down each column from fig1 above
Also creates L variable
'''
def log10mean(X):
    L=10* np.log10(np.mean((10**(X/10))))
    return L

Snum=np.reshape(np.arange(50,650,50),(12,1))
L=Snum+3*data_array1  

YS21_ave=np.zeros((12,6))

for a in np.arange(0,6):
    for b in np.arange(0,12):
        c=a*12+b
        YS21_ave[b,a]=log10mean(img_YS21[:,c])
#         print( a, b, c)
```

```python
## Plots Ave S21

'''Generates plot of YS21-ave vs L'''

'''Sets up plot parameters and generates plot'''
plt.rc('font', weight='bold')
plt.rcParams['axes.linewidth']=4
plt.rcParams['figure.figsize'] = (16,9)

fig2 = plt.figure()

plt.tick_params(axis='both', which='major', labelsize=24, width=4, length=8, labeltop=False)
plt.tick_params(axis='both', which='minor', labelsize=24, width=4, length=8, labeltop=False)

plt.xlabel(r'CPW Width, L ($\mu$m)', fontsize=36, fontweight='bold')
plt.ylabel(r'$S_{21-ave}$ (GHz)', fontsize=36, fontweight='bold')

plt.xlim(50,625)
plt.ylim(-.75,-.25)

'''Sets up colorscheme'''
jet = plt.get_cmap(cm.jet_r)
cNorm  = colors.Normalize(vmin=2, vmax=13.1)
scalarMap = cm.ScalarMappable(norm=cNorm, cmap=jet)
values=[2,3.8,4.4,6.7,10,13.1]
labels = ['PTFE','Quartz','FR-4','Beryllia','Alumina', 'GaAs']

plotH= []
for i in np.arange(0, 6):
    
    colorVal = scalarMap.to_rgba(values[i])
    x,=plt.plot(L[:,i],YS21_ave[:,i],color=colorVal, linewidth=8, label=labels[i])

'''Generates Legend'''
leg=plt.legend(ncol=2,fontsize=30, loc=8, frameon=False, handlelength=2)

for legobj in leg.legendHandles:
    legobj.set_linewidth(8.0)
    
```

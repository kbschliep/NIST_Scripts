```python
## Load Data  use for .csv data
''' Open data and set up for analysis'''

num_files = np.size(file_list)-1 # number of files in the directory with the correct file type
file_shape = np.array(pd.read_csv(file_list[0]).values.shape) # data file size
matrix_shape = np.append(file_shape, num_files) # size of matrix to hold all data
data_array = np.empty(matrix_shape) # builds empty matrix of size matrix_shape
a=0
for fname in file_list: #iterates through files along order of glob.glob(path)

    if fname==file_list[1]:
        continue
    else:
        data = pd.read_csv(fname) # uses pandas to open files and keep structure
        datapd = data.values # converts from dataframe to numpy array
#         print(a)
        data_array[:,:,a] = datapd # inserts each file as new layer of matrix along z axis (axis=2)
        a+=1 # increase counter to progress through list
    

    
```

```python
## Reorganize Data into Correct Order and 
X=data_array[:,0,0]; # All used the same x range
data_array_reorder=np.zeros(data_array.shape)
listr=[1,2,4,5,6,3]
a=0
for n in listr:
    data_array_reorder[:,:,int(n)-1]=data_array[:,:,a];# puts array in order of er ->2,3.8,4.4,6.7 10, 13.1
    
    a+=1

    
    
YS21=data_array_reorder[:,2::2,:].copy();# extracts S21 from each file - wrong for 3 aka er=6.7 data_array[:,:,2]
YS11=data_array_reorder[:,1::2,:].copy();# extracts S11 from each file - wrong for 3 aka er=6.7 data_array[:,:,2]
YS21[:,:,3]=data_array_reorder[:,13:,3]
YS11[:,:,3]=data_array_reorder[:,1:13,3]
```

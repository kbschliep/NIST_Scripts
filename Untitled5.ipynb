```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal as ss
```

```python
def normalize(raw):
    norm=[]
    for i in raw:
        if i<0:
            norm.append(-i/min(raw))
        elif i>0:
            norm.append(i/max(raw))
        else:
            norm.append(i)
    return norm
# sample rate is 0.01 seconds
dt = 0.001
t = np.arange(0,1,dt)
f1=10
f2=100
B=f1*2
f = np.sin(2*np.pi*t*f1)+np.sin(2*np.pi*t*f2)
h = np.sinc(2*B*t)
c = dt*np.convolve(f,h)
c2 = dt*np.convolve(c,h)
# sample rate is now 0.001 seconds
#dt2 = 0.001
#t2 = np.arange(0,1,dt2)
#f2 = np.sin(2*np.pi*t2*f1)+np.sin(2*np.pi*t2*f2)
#h2 = np.sinc(2*B*t2)
#c2 = dt2*np.convolve(f2,h2)
# confirm that the two convolutions are the same
plt.plot(t,normalize(c[0:t.shape[0]]),'b-')
plt.plot(t2,normalize(c2[0:t2.shape[0]]),'g-')
plt.plot(t,normalize(f),'r-')
#plt.plot(t2,normalize(f2),'y-')
#plt.ylim(ymax=1500)  # adjust the max leaving min unchanged
#plt.ylim(ymin=-1500)  # adjust the min leaving max unchanged
plt.xlim(xmax=.2)  # adjust the max leaving min unchanged
plt.xlim(xmin=0)  # adjust the min leaving max unchanged
```

```python
cfft=np.fft.rfft(c)
c2fft=np.fft.rfft(c2)
ffft=np.fft.rfft(f)

plt.plot(normalize(cfft[0:200]),'b-')
plt.plot(normalize(c2fft[0:200]),'g-')
plt.plot(normalize(ffft[0:200]),'r-')
```

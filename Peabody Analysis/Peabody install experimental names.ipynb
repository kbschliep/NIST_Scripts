```python
A='PO'
B='NPA'
C='OB'
D='PA'
E='P'
F='KP'
G='CA'
H='BC'
I='F'
a='-'
J='?'
```

```python
ls_1=[]
num=0
num_BC=9
num_CA=4
num_PA=4


for j in range(num_BC):
    for i in range(num_CA):
        ls_1.append(A+a+H+str(j+1)+a+B+a+G+str(i+1))
        num+=1
    ls_1.append(A+a+H+str(j+1)+a+B+a+G+J+a+C)

for k in range(num_PA):
    for j in range(num_BC):
        for i in range(num_CA):
            ls_1.append(A+a+H+str(j+1)+a+D+str(k+1)+a+G+str(i+1))
            num+=1
        ls_1.append(A+a+H+str(j+1)+a+D+str(k+1)+a+G+J+a+C)
freq=['2','3','4','5','6']
powers=['10W','20W','30W' ]
for m in powers:
    for l in freq:
        for j in range(num_BC):
            for i in range(num_CA):
                ls_1.append(E+a+F+m+a+I+l+a+H+str(j+1)+a+B+a+G+str(i+1))
                num+=1
            ls_1.append(E+a+F+m+a+I+l+a+H+str(j+1)+a+B+a+G+J+a+C)

        for k in range(num_PA):
            for j in range(num_BC):
                for i in range(num_CA):
                    ls_1.append(E+a+F+m+a+I+l+a+H+str(j+1)+a+D+str(k+1)+a+G+str(i+1))
                    num+=1
                ls_1.append(E+a+F+m+a+I+l+a+H+str(j+1)+a+D+str(k+1)+a+G+J+a+C)
```
